```python
from src.data_processing import HyperspectralImage, OriginalImageDataset
from torch.utils.data import DataLoader
from src.models import Autoencoder
from src.train_utils import train_model
from src.conf import BATCH_SIZE
from src.loss_functions import sad_loss
import matplotlib.pyplot as plt
from src.metrics import compare_abundances, compare_endmembers
import torch
```

```python
urban = HyperspectralImage(image_path="datasets/urban/Urban.mat",
                                ground_truth_path="datasets/urban/end6_groundTruth.mat",
                                n_bands=162, n_row=307, n_col=307, n_sources=6)

plt.imshow(urban.X[0])
plt.show()

original_image_dataset = OriginalImageDataset(hsi=urban, with_patches=False)
print(original_image_dataset.X.shape)
plt.imshow(original_image_dataset.X[0])
plt.show()



original_image_dataloader = DataLoader(dataset=original_image_dataset, batch_size=10, shuffle=True)
a = next(iter(original_image_dataloader))
print('original_image_dataloader', a["X"].shape)
plt.imshow(a['X'][0][0])
plt.show()
```

```python
model = Autoencoder(channel_number=urban.n_bands, source_number=urban.n_sources)
model, loss_train = train_model(model, urban.X,
                                              original_image_dataloader,
                                              sad_loss, num_epochs=100, n_sources=6, need_plot=True)
```

```python
torch.save(model.state_dict(), "results/models_weights/urban_original_no_patches.pth")

X, A, S = model(torch.Tensor(urban.X[None, :, :, :]))
X, A, S = X.detach().numpy()[0], A.detach().numpy(), S.detach().numpy()[0]
# Endmembers normalization
for i in range(A.shape[1]):
        M = A[:, i].max().item()
        A[:, i] = A[:, i]/M
                      
print('compare_endmembers', compare_endmembers(urban, A))
plt.figure(figsize=(8,6))
for ii in range(6): 
    ax = plt.subplot(2, 3, ii+1)
    ax.plot(A[:,ii])
plt.show()
                
print('compare_abundances', compare_abundances(urban, S))
plt.figure(figsize=(8,6))
for ii in range(6): 
    ax = plt.subplot(2, 3, ii+1)
    ax.imshow(S[ii, :, :])  
plt.show()
```
